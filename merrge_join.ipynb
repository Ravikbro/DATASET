```python
import numpy as np
import pandas as pd
```

```python
A  = pd.DataFrame({
    "Transport":["Poyezd", "Avtomobil", "Kema"],
    "Narxi":["98000","10000", "65000"]
})
A
```

```python
B=pd.DataFrame({
    "Transport":["Poyezd", "Avtomobil", "Tramvay"],
    "Tezligi":["70","80","145"]
})
B
```

```python
merge = pd.merge(B,A, how="inner")#inner join
merge
```

```python
merged_outer = pd.merge(A,B,how="outer")#outer join
merged_outer
```

```python
merge_left = pd.merge(A,B, how="left") #left join
merge_left
```

```python
merge_right = pd.merge(A,B,how="right")
```
